**Notebook of quantum gates and circuits with IBM Qiskit**

```python
# External dependencies

from qiskit.circuit import QuantumCircuit
import numpy as np
```

```python
# Parameterized RX Quantum Gate

theta = 0.2  # Parameter
circuit = QuantumCircuit(1)
circuit.rx(theta, 0)  # RX gate with parameter in qubit 0

circuit.draw(output='mpl')
```

```python
# Qiskit Euler single-qubit rotation

from qiskit.circuit.library import UGate

# u(theta, phi, lam, qubit)
circuit = QuantumCircuit(1)
circuit.u(np.pi, np.pi/2, np.pi/4, 0)  # RX gate with parameter in qubit 0

circuit.draw(output='mpl')
```
