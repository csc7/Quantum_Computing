### Quantum Fourier Transform

```python
# Qiskit Libraries

from qiskit import *
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
# Definition of classical and quantum bits

quantum_bits = QuantumRegister(4)
classical_bits = ClassicalRegister(4)

circuit = QuantumCircuit(quantum_bits, classical_bits)
```

```python
# Circuit

circuit.h(2)
circuit.cs(3, 2)
circuit.h(3)
circuit.swap(3, 2)

circuit.draw(output='mpl')
```

```python
# Circuit measurement

circuit.measure([0, 1, 2, 3], [0, 1, 2, 3])
```

```python
# Simulation
simulator = AerSimulator()

# decompose solution in:
# https://quantumcomputing.stackexchange.com/questions/34396/aererror-unknown-instruction-c-unitary-while-using-control-unitary-operator

# Result and Plot
result = simulator.run(circuit.decompose(reps=6)).result()
counts = result.get_counts(circuit)
plot_histogram(counts, title='Bell-State counts')
```
